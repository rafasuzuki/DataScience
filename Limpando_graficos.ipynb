# Limpando Gráficos

Gráficos e números tem sua correlação. O que seria um gráfico se não uma representação de dados e informações numéricas. Eles estão por todos os lados. Quando você assiste o noticiário ou lê um jornal, alta do dólar, inflação, dados do Covid e diversas outras representações númericas comparativas em imagens.
Você, assim como eu, fica perdido na quantidade de informação naquela quantidade de linhas, barras e números?
Nesse trabalho quero te apresentar o príncipio da Gestalt, e como podemos limpar nossos gráficos os tornando mais fácil e acessível a todas as pessoas. 


##Importando Bibliotecas

Importarei a biblioteca `numpy` e para a visualização dos nosso gráfico a biblioteca `matplotlib`.

```python
import matplotlib.pyplot as plt
import numpy as np
```

#Ajuste 1
##Eliminando bordas, principio Gestalt e setando limite do eixo Y. Linhas mais grossas com (linewidth).


A **Gestalt** afirma que sempre compreendemos o todo de um composto visual, antes de aprofundar nos seus detalhes. Dentro do compilado, há elementos que seu cérebro consegue destrinchar com rapidez e outros mais complexos que você não associa tão rapidamente.

O nível de facilidade com que identificamos e compreendemos visualmente uma peça é mensurado através da pregnância. Dado esse fato, a pregnância de um objeto é que vai definir o quão rápido ele é percebido e assimilado, e também a eficiência da comunicação da mensagem para com o receptor.

Logo, queremos nosso gráfico o mais limpo possível. Para isso, removeremos detalhes desnecessários e realçaremos outros que merecem nossa atenção.

```python
#criando objeto
fig, ax = plt.subplots(figsize = (10,6), dpi=72)

#removendo bordas desnecessarias
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

#definindo as linhas
ax.plot(range(12),alunos , color = VERMELHO, linewidth = 5)
ax.plot(range(12),projetos, color = CINZA, linewidth = 5)

#setando limite do eixo Y
ax.set_ylim(0,300)

#exibindo o grafico
plt.tight_layout()
```

Acima, removemos as bordas opostas aos nosso eixos e realçamos as linhas que representam nossas informações relevantes.
Aproveitei para setar nosso eixo y iniciando em 0, antes ele começava do 120.

#Ajuste 2
## Removendo yticks e ytickslabel, renomeando xticklabels

Como disse antes, queremos um gráfico limpo e de facil compreensão. Então não necessitamos de todos esse numeros em nossos eixos. Para o eixo Y, removi os números e mais a frente colocarei os valores nas próprias linhas, eliminando informações desnecessárias. Para nosso eixo X, como os dados fictícios representam ao longo de um ano, faz mais sentido inserir os meses correspondentes.

```python
#criando objeto
fig, ax = plt.subplots(figsize = (10,6), dpi=72)

#removendo bordas desnecessarias
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

#definindo as linhas
ax.plot(range(12),alunos , color = VERMELHO, linewidth = 5)
ax.plot(range(12),projetos, color = CINZA, linewidth = 5)

#setando limite do eixo Y
ax.set_ylim(0,300)

#removendo yticks e yticklabel
ax.set_yticks(ticks=[])
ax.set_yticklabels(labels=[], visible=False)

#renomeando xticklabels
ax.set_xticks(range(12))
ax.set_xticklabels(labels=X)


#exibindo o grafico
plt.tight_layout()
```

#Ajuste 3
## Plotando o scatterplot e mostrando os valores da linha, as 3 últimas.
Nessa fase de limpeza do nosso gráfico, decidi por colocar as informações correspondentes aos 3 últimos meses. Não só por ser os últimos meses do ano, mas também por serem os meses com maior diferença entre as ligações e casos resolvidos.

```python
#criando objeto
fig, ax = plt.subplots(figsize = (10,6), dpi=72)

#removendo bordas desnecessarias
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

#definindo as linhas
ax.plot(range(12),alunos , color = VERMELHO, linewidth = 5)
ax.plot(range(12),projetos, color = CINZA, linewidth = 5)

#plotando scatterplot #clip_on, caso o ultimo scatter fique pela metade
ax.scatter(range(12)[-3:], alunos[-3:], s=95, color = VERMELHO, clip_on=False)
ax.scatter(range(12)[-3:], projetos[-3:], s=95, color = CINZA, clip_on=False)

#anotacoes nos pontos do scatter
for i, j in enumerate(alunos[-3:]):
  ax.annotate(
      j,
      xy=(i+9, j),
      xytext=(5,5),
      textcoords='offset points',
      color=VERMELHO,
      fontsize=10,
      fontweight='medium'
  )

for i, j in enumerate(projetos[-3:]):
  ax.annotate(
      j,
      xy=(i+9, j),
      xytext=(5,5),
      textcoords='offset points',
      color=VERDE,
      fontsize=10,
      fontweight='medium'
  )

#setando limite do eixo Y
ax.set_ylim(0,300)

#removendo yticks e yticklabel
ax.set_yticks(ticks=[])
ax.set_yticklabels(labels=[], visible=False)

#renomeando xticklabels
ax.set_xticks(range(12))
ax.set_xticklabels(labels=X)


#exibindo o grafico
plt.tight_layout()
```

#Ajuste 4
##Dando nome as linhas. Titulos, Subtitulos e Rodapé.

Com nosso gráfico limpo, podemos inserir as informações pertinentes a ele.
